```python
import requests
from bs4 import BeautifulSoup

def get_video_source(url):
    try:
        # Send GET request
        response = requests.get(url)
        response.raise_for_status()
        
        # Parse HTML
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find video element
        video_element = soup.find('video')

        print(soup)
        
        if video_element and video_element.find('source'):
            source_url = video_element.find('source')['src']
            print(f"Video source URL: {source_url}")
            return source_url
        else:
            print("No video source found on the page")
            return None
            
    except requests.RequestException as e:
        print(f"Error fetching the page: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

url = "https://www.ubu.com/film/rosefeldt_asylum.html"
source = get_video_source(url)
```

```python
import subprocess
import os, pathlib
# Define the output directory and file name
output_dir = os.path.expanduser("~/A Video")
output_file = pathlib.Path(source).name
output_path = os.path.join(output_dir, output_file)

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)
print(output_path)
# Use subprocess to run the curl command
subprocess.run(["curl", "-L", source, "-o", output_path])
```
